```python
import numpy as np
import matplotlib.pyplot as plt

# 
radius = 5
steps = 100
velocity = 1
height = 10
turns = 3

# 8
figure_eight_x = []
figure_eight_y = []
for i in range(steps):
    angle = 2 * np.pi * (i / steps)
    x = radius * np.sin(angle)
    y = radius * np.sin(2 * angle) * 0.5
    figure_eight_x.append(x)
    figure_eight_y.append(y)

# 
spiral_x = []
spiral_y = []
spiral_z = []
for i in range(steps):
    angle = 2 * np.pi * turns * (i / steps)
    x = radius * np.cos(angle)
    y = radius * np.sin(angle)
    z = (height / steps) * i
    spiral_x.append(x)
    spiral_y.append(y)
    spiral_z.append(z)

# 8
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(figure_eight_x, figure_eight_y)
plt.title('Figure Eight Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')

# 
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot(spiral_x, spiral_y, spiral_z)
ax.set_title('Spiral Ascent Curve')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
radius = 5
steps = 100

# 
theta = np.linspace(0, 2 * np.pi, steps)
circle1_x = radius * np.cos(theta)
circle1_y = radius * np.sin(theta)

# 
circle2_x = radius * np.cos(theta) + 2 * radius
circle2_y = radius * np.sin(theta)

# 8
plt.figure(figsize=(6, 6))
plt.plot(circle1_x, circle1_y, label='Circle 1')
plt.plot(circle2_x, circle2_y, label='Circle 2')
plt.title('Figure Eight Curve (Two Circles)')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.legend()
plt.show()
```
